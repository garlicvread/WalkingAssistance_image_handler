```python
# Importing all necessary libraries
import cv2 as cv
import os
import time
```

```python
# Get the number of the files in the directory
path = "/home/raymond/Downloads/development/imageExtractor/videos"
train_path = "/home/raymond/Downloads/development/imageExtractor/train"
test_path = "/home/raymond/Downloads/development/imageExtractor/test"

file_list = os.listdir(path)

print(file_list)
print("Number of files in path: ", len(file_list) - 1)
```

```python
count = 1
train = 1
test = 1

start_time = time.time()
print("\n=== Start image extraction process ===\n")

# Read the video from specified path
for i in range(1, len(file_list)):
    image_capture = cv.VideoCapture(f"{path}/%d.mp4" %i)
    
    print("\n=== Start reading %d.mp4 ===\n" %i)
    
    try:
        # Creat a directory named 'train'
        if not os.path.exists('train'):
            os.makedirs('train')
        
        # Create a directory named 'test'
        if not os.path.exists('test'):
            os.makedirs('test')

    # If the program does not creat the directory then raise an error
    except OSError:
        print ('Error: Creating directories')

    # Declare current frame index
    current_frame = 0

    while(True):
        # Read image from each frame
        ret, frame = image_capture.read()

        if ret:
            if (int(image_capture.get(1) % 10 == 0)):

                # writing the extracted images
                cv.imwrite(f"{test_path}/test_%s.jpg" %str(test).zfill(7), frame)
                
                # Print an image file extraction message
                file_name = str(test).zfill(7)
                print('Extracting test' + f'{file_name}.jpg')
                
                # Increasing the counter so that it will show how many frames are created
                test += 1
            
            else:
                # writing the extracted images
                cv.imwrite(f"{train_path}/train_%s.jpg" %str(train).zfill(7), frame)
                
                # Print an image file extraction message
                file_name = str(train).zfill(7)
                print('Extracting train' + f'{file_name}.jpg')
                
                # Increasing the counter so that it will show how many frames are created
                train += 1
            
        else:
            break
            
    # Release all space and windows once done
    image_capture.release()
    
    print("\n=== End reading %d.mp4 ===\n" %i)
    
    cv.destroyAllWindows()
    
end_time = time.time()

print('\n=== End image extraction process ===')
print("\n\nTotal processing time:", round((end_time - start_time), 2), "seconds")
```
